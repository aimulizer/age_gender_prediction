```python
import cv2
import tensorflow.keras.models as  ld
```

```python
gender_model=ld.load_model("genderfinal.model")
```

```python
MODEL_MEAN_VALUES = (78.4263377603, 87.7689143744, 114.895847746)
age_list = ['(0, 2)', '(4, 6)', '(8, 12)', '(15, 20)', '(25, 32)', '(38, 43)', '(48, 53)', '(60, 100)']
age_net = cv2.dnn.readNetFromCaffe('deploy_age.prototxt', 'age_net.caffemodel')
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
classifier=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
gender_list=['male','female']
import numpy as np
```

```python
video=cv2.VideoCapture(0)
while True:
    ret,frame = video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    face = classifier.detectMultiScale(gray,1.3,5)
    
    for x,y,w,h in face:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,255,0),3)
        new_array=cv2.resize(gray[y:y+h,x:x+w],(70,70))
        new_array=new_array.reshape(-1,70,70,1)
        
        ypred=gender_model.predict(new_array)
#         print(ypred[[0]])
        ypred = ypred[[0]]
        if ypred>0.5:
            ypred=1
        else:
            ypred=0
        gender=gender_list[ypred]
        
        
        #Get Face 
        face_img = frame[y:y+h, x:x+w].copy()
        blob = cv2.dnn.blobFromImage(face_img, 1, (227, 227), MODEL_MEAN_VALUES, swapRB=False)
        
        age_net.setInput(blob)
        age_preds = age_net.forward()
        age = age_list[age_preds[0].argmax()]
        print("Age Range: " + age)
        overlay_text = "%s %s" % (gender, age)
        cv2.putText(frame, overlay_text, (x, y), font, 1, (255, 255, 255), 2, cv2.LINE_AA)
        
        
    cv2.imshow("legend",frame)
    
    if cv2.waitKey(1) & 0xff==ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
